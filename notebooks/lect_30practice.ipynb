```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import boxcox
```

```python
df = pd.read_csv("Indian Liver Patient Dataset (ILPD).csv")
df.head()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(11,3))
plt.subplot(121)
sns.histplot(df['direct_bilirubin'],kde=True,bins=50)
plt.title('tot_bilirubin')

plt.subplot(122)
stats.probplot(df['direct_bilirubin'],dist='norm',plot=plt)
plt.title('QQ plot tot_bilirubin')

print(df[['direct_bilirubin']].skew())


from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer(method='box-cox')
df['direct_bilirubin'] = pt.fit_transform(df[['direct_bilirubin']])

print('after transformation ',df[['direct_bilirubin']].skew())


plt.figure(figsize=(11,3))
plt.subplot(121)
sns.histplot(df['direct_bilirubin'],kde=True,bins=50)
plt.title('tot_bilirubin')

plt.subplot(122)
stats.probplot(df['direct_bilirubin'],dist='norm',plot=plt)
plt.title('QQ plot tot_bilirubin')

plt.show()
```
