```python
np.random.seed(42)
num_sample = 50
dataset_features = {
    'age': {'type': 'integer', 'min': 18, 'max': 65},
    'income': {'type': 'float', 'min': 2000, 'max': 15000},
    'height': {'type': 'float', 'min': 1.5, 'max': 2.0},
    'city': {'type': 'categorical', 'choices': ['New York', 'London', 'Paris', 'Tokyo']},
    'score': {'type': 'float', 'min': 0, 'max': 100}
}

data = {}

for feature_name, parameters in dataset_features.items():
    if parameters['type'] == 'integer':
        data[feature_name] = np.random.randint(parameters['min'], parameters['max'] + 1, num_sample)
    elif parameters['type'] == 'float':
        data[feature_name] = np.random.uniform(parameters['min'], parameters['max'], num_sample)
    elif parameters['type'] == 'categorical':
        data[feature_name] = np.random.choice(parameters['choices'], num_sample)
data
```
