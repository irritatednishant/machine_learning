```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('Vehicle Price.csv')
df.head()
```

```python
df['make'].value_counts()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df[['make','year','price','fuel','cylinders','mileage']]
```

```python
df.sample(5)
```

```python
# Missing value treatment

df.dropna(subset=['price','fuel'],inplace=True)

trf1 = ColumnTransformer([
    ('simple_imp',SimpleImputer(strategy='mean'),['cylinders','mileage'])
],remainder='passthrough')
```

```python
trf2 = ColumnTransformer([
    ('ohe',OneHotEncoder(drop='first',sparse_output=False),['make','price'])
],remainder='passthrough')
```

```python
trf3 = LinearRegression()
```

```python
x_train,x_test,y_train,y_test= train_test_split(df[['make','year','fuel','cylinders','mileage']],df[['price']], test_size=0.2,random_state=42)
```

```python
x_train
```

```python
pipe = Pipeline([
   ( 'trf1',trf1),
   ( 'trf2',trf2),
   ( 'trf3',trf3)
])
```

```python
pipe.fit(x_train,y_train)
```
